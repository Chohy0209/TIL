```python
#knn알고리즘 활용해서 test survived알아보기
#test데이터의 사람과 유클리디안 거리상 가장 가까운 train 데이터의 survived를 k개 추출해서 다수결로 정하기
from sklearn.neighbors import KNeighborsClassifier
```

```python
clf = KNeighborsClassifier(n_neighbors =5) #k=13
```

```python
from sklearn.model_selection import KFold
from sklearn.model_selection import cross_val_score
k_fold = KFold(n_splits=10, shuffle=True, random_state=0) #k-fold 검증 중 fold=10개 
```

```python
scoring = 'accuracy'
score = cross_val_score(clf, train_data, target, cv=k_fold, n_jobs=1, scoring=scoring)
print(score)
```

```python
# clf= k가 13인 knn모델
# train_data에 저장된 8개의 열에대해 target에 clf로 10개의-fold로 검증한 정확도
train_data
```

```python
round(np.mean(score)*100,2)
```
