```python
import pandas as pd
import numpy as np
```

```python
#협업필터링
movie_data=pd.read_csv("../../data/archive/movies.csv")
movie_data # 영화 데이터
```

```python
rating_data=pd.read_csv("../../data/archive/ratings.csv")
rating_data #사용자-평점 데이터
```

```python
rating_data.drop('timestamp',axis=1,inplace=True)
rating_data.head(2)
```

```python
user_movie_rating=pd.merge(rating_data,movie_data,on='movieId')
user_movie_rating.head(2)
```

```python
movie_user_rating=user_movie_rating.pivot_table('rating', index='title', columns='userId')
movie_user_rating.head(5)
```

```python
user_movie_rating=user_movie_rating.pivot_table('rating', index='userId', columns='title')#.T를 이용해서도 이렇게 할 수 있음.
user_movie_rating.head(5)
```

```python
movie_user_rating.head(5)
#아이템 기반 협업 필터링 : 어떤 영화에 대한 평가 점수 분포와 가장 비슷한 평가분포를 가진 영화를 추천
#nan을 0으로 대체
movie_user_rating.fillna(0,inplace=True)
movie_user_rating.head()
```

```python
m1=np.array([0,1,1])
m2=np.array([1,0,1])
m3=np.array([2,2,2])
m4=np.array([1,1,1])
```

```python
from numpy.linalg import norm #벡터의 크기를 구하는 모듈
def cossim(a,b):
    return np.dot(a,b)/(norm(a)*norm(b)) # 벡터의 내적을 두벡터 크기의 곱으로 나눔
print(cossim(m1,m4))
print(cossim(m2,m4))
print(cossim(m3,m4))
#코사인 유사도
```

```python
from sklearn.metrics.pairwise import cosine_similarity #코사인 유사도를 구해주는 함수
item_based_cola_sim=cosine_similarity(movie_user_rating)
item_based_cola_sim
```

```python
pd.DataFrame(data=item_based_cola_sim,index=movie_user_rating.index,columns=movie_user_rating.index)["user_movie"].sort_values(ascending=False).index[1]
```

```python
#아이템 기반 영화 추천 코사인유사도 이용 5가지
user_movie=input("영화를 입력하세요")
" ".join(pd.DataFrame(data=item_based_cola_sim,index=movie_user_rating.index,columns=movie_user_rating.index)[user_movie].sort_values(ascending=False).index[1:6])
```
