```python
#(정규식)(정규식)
#그룹: (정규식)
res=re.match("([0-9]+) ([0-9]+)", "100 200 300") # 100공백200 까지 매칭
res
```

```python
res.group(1) #매칭된 첫번째 그룹
```

```python
res.group() #매칭된 문자열 전체
```

```python
res.groups() #패턴에 매치된 문자열을 튜플형식으로 리턴
```

```python
#패턴 그룹에 대한 이름 부여
#(?P<그룹이름>정규식) 패턴식에 소괄호를 묶으면 그룹
res=re.match("(?P<func>[a-z]+)","print(100)")
res
```

```python
re.sub("\d+", "num", "10 20 sky 50 100 bird 200")
#문자열에서 숫자는 모두 2배를 적용하시오.
#'20 40 sky 100 200 bird 400"
# 변경함수(매치된 객체):
#        변경할 수식
#re.sub("패턴",변경함수,"문자열","변경횟수")

def func_db(res):
    #변경할 수식
    return str(int(res.group())*2)
re.sub("\d+",func_db,"20 40 sky 100 200 bird 400") #매치된 모든 객체를 func_db에 호출해라 매치가 안된건 따로 문자열로 분류되서 남음
```

```python
test="1 2 3 abc 4 def"
#문자열중 숫자를 모두 10배 하시오
def func_db(res):
    return str(int(res.group())*10)
re.sub("\d+",func_db,test)
```

```python
import re
emails = ['python@mail.example.com', 'python+kr@example.com',              # 올바른 형식
          'python-test@example.co.kr', 'python_10@example.info',         # 올바른 형식
          'python.test@e-xample.com',                                    # 올바른 형식
          '@example.com', 'python@example', 'python@example-com']          # 잘못된 형식
a=[]
for i in emails:
    res = re.match("[A-z+-_.]+@[A-z.-]+[.][A-z]+", i)
    if res:
        a.append("올바른 형식")
    else:
        a.append("잘못된 형식")
res=' '.join(a)
print(res)
```

```python
import random
x=random.randint(1,100)
def Numgame():
        count=0
        while True:

            user=int(input("숫자를 입력하세요"))
            if user>x:
                print("DOWN")
                count+=1

            elif user<x:
                print("up")
                count+=1
            elif user==x:
                print("정답입니다")
                return count
                break
a=Numgame()
print(a+1,"번만에 정답을 맞췄습니다")
```

```python
import math
x,y=map(int,input("밑변과 높이길이를 써주세요").split())
def Pita(x,y):
    C=(x**2)+(y**2)
    c=math.sqrt(C)
    return c
Z=Pita(x,y)
print(Z)
```
