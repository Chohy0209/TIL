```python
#ex6.csv key열에 대해 문자별 빈도수를 출력
chker=pd.read_csv("../../data/examples/ex6.csv",chunksize=1000)
type(chker)
tot=pd.Series([],dtype='int64')
for piece in chker: #데이터를 chunksize인 1000개씩 짤라서 1000개 단위로 나눠서 piece로 들어가서 for문으로 작업
    x=piece['key'].value_counts()
    tot=tot.add(x,fill_value=0)
tot.loc['A':'Z']
```

```python
chker=pd.read_csv("../../data/examples/ex6.csv",chunksize=1000)
type(chker)
tot=pd.Series([],dtype='int64')
for piece in chker:
    x=piece['key'].value_counts()
    tot=tot.add(x,fill_value=0)
tot.loc['A':'Z']
```
