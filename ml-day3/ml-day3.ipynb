```python
#범주형 결측값 대체방법 : 그룹화 후 최빈값으로 대체
#연속형 결측값 대체방법 :  그룹화 후 통계값으로 대체
```

```python
#transform 함수 용도: 그룹별로 계산된 통계값으로 특정값을 대체할때 유용
```

```python
#df의 Fruit열 결측값을 Gender 그룹의 최빈값으로 대체하기
df['Fruit'].fillna(df.groupby('Gender')['Fruit'].transform(lambda x:x.value_counts().idxmax()),inplace=True)
df
#x에는 Gender별 Fruit열이 전달됨
```

```python
df['Gender'].value_counts().idxmax() #idxmax = 시리즈에서 최빈값의 인덱스를 추출해주는 함수
```

```python
#df의 height열 대체 결측값을 그룹별 평균으로 대체

df['Height'].fillna(df.groupby('Gender')['Height'].transform(lambda x:x.mean()),inplace=True)
df
```
