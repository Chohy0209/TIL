```python
def myTop3(df):
    return df.sort_values(by="petal_length",ascending=False)[:3] #by petal_length로 기준을 정하고 return을 하면 시리즈가 데이터프레임형식으로 넘어감
    #print(df.sort_values(by="petal_length",ascending=False)[:3])
    #print(df)
    #print("="*50)
iris.groupby(iris.species).apply(myTop3)
```

```python
#transform: 그룹화 연산에 따른 데이터프레임의 변환
iris.groupby(iris.species)
```

```python
def q3(s):
    return pd.cut(s,3,labels=['소','중','대'])
    
iris['p_l_class']=iris.groupby(iris.species).petal_length.transform(q3) # transform 각각의 값에 대해 시행
#species를 기준으로 그루핑하여 
#각 그룹에 대한 꽃잎 길이를 기준으로 3개의 서브그룹으로 나누고 각각 소,중,대 라벨 부여
```

```python
iris[['petal_length','p_l_class']] #종별 petal_length에 따라 다시 3개의 구간으로 cut
#특성공학 (feature engineering): 데이터프레임의 열에대한 가공처리를 한 결과로 새로운 열을 생성(추가)
```

```python
#pivot_table : pivot(피봇 테이블) + groupby(그룹분석)
#df1.pivot(행,열,데이터)
#df1.pivot_table(데이터,행,열)
df1.pivot_table('인구','도시','연도')
df1.pivot_table(values='인구',columns='연도',index='도시',margins=True) 
#margins 옵션을 True로 해서 각각의 그룹별 특정계산(aggfunc : 애그리게이트펑션) 을 실행해서 보여줌 aggfunc의 default값은 mean(평균)
```
