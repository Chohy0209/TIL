```python
# 연관규칙(Association Rules): 어떤 물건(item)을 샀을때 함께 구매할것으로 예측되는 규칙 {x}->{y} x를 산사람이 y까지 같이 산다고 예측되는 규칙을 찾음
#  트랜잭션(Transaction): 특정 시간 동안 발생한 아이템들의 집합. {빵, 우유} <- 한 트랜잭션

# Support (지지도 S) : 특정 아이템셋을 포함하는 트랜잭션이 전체에서 몇건이 있나 비율 -> 총 5건의 트랜잭션 중 특정아이템셋이 2번이었다면 support= 2/5
# Frequent itemset(빈발아이템집합,빈발셋) : 지지도가 최소 지지도에대한 임계값(하이퍼 파라미터)보다 크거나 같은경우 
```

```python
# S(x->y): x와 y를 모두 포함하는 거래의 비율. x를 구매한사람이 y를 구매했다라는 연관규칙의 지지도. 분모가 total
# Confidence(신뢰도 C): 
# C(x->y): x아이템 셋을 포함하는 거래들 중 y 아이템셋의 출현 빈도 분모가 아이템셋 x 의 빈도
```

```python
#하이퍼 파라미터 설정 confidence>=minconf threshold
```

```python
# 연관규칙 단계
# 1. Frequent itemset generation (minsup 하이퍼 파라미터 설정) 빈발셋 생성
# 2. 빈발셋 중 컨피던스가 높은 규칙 생성
# 연관규칙 단점 -> 컴퓨팅 계산이 너무 크게 일어남
#5개 물품의 규칙 전체 개수 = 3**5- 2**(5+1)+1
```

```python
# Apriori 알고리즘
# {c,d,e}가 Frequent itemset 이라면 {c},{d},{e}도 빈발셋
# -> {A,B}가 비빈발항목 infrequent라면 그것을 포함하는 하위 항목도 전부 비빈발항목이므로 계산 x
```

```python
# lift(향상도): {x->y}의 향상도 = x를 산 사람중 y를 산 사람의 확률(confidence(x->y))/전체중에 Y를 살 확률 (surpport(Y))
# 전체중에 y를 사는사람과 x중 y를 사는 사람들간의 비율 1보다 높을수록 x와 y의 연관규칙 관계성이 높다
```

```python
from mlxtend.frequent_patterns import apriori,association_rules
```

```python
df=pd.read_csv("../../data/data/GroceryStoreDataSet.csv",names=['products'])
u=df['products'].apply(lambda x:x.split(","))
u
```

```python
data=list(df['products'].apply(lambda x:x.split(","))) #apriori를 돌릴때 데이터를 2중 리스트 구조로 변환해야함,각각의 리스트가 트랜잭션에 해당
data
```

```python
from mlxtend.preprocessing import TransactionEncoder #2중리스트 구조를 트랜잭션 인코더로 변환
```

```python
te=TransactionEncoder()
```

```python
trans_data=te.fit(data).transform(data)
trans_data
```

```python
te.columns_ #트랜잭션 인코딩 순서 
```

```python

df=pd.DataFrame(trans_data,columns=te.columns_) #여기까지 만들어 놔야 연관규칙 작업준비 완료.
df
```

```python
#트랜잭션 데이터프레임-> apriori -> association rules ->lift -> 규칙 추출
df=apriori(df,min_support=0.2,use_colnames=True)# 최소 지지도 0.2 이상인 아이템셋 추출
df
```

```python
df_ar=association_rules(df,metric="confidence",min_threshold=0.6)
df_ar
#{milk->bread}  confidence :밀크를 산 사람중에 80퍼는 빵을 삼/  
#lift:전체고객중 빵을 구매한 고객비율보다 밀크를 산 사람중 빵을 구매한 고객비율이 1.2배 더 높음
```

```python
df_ar.sort_values('lift',ascending=False)[['antecedents','consequents']]
```
