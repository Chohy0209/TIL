```python
test_df=pd.read_csv("../../data/titanic/test.csv")

test_df['Survived']=0
test_df.loc[(test_df['Sex'] == 'female') & ((test_df['Pclass'] == 1)|(test_df['Pclass'] == 1)), 'Survived'] = 1
test_df.loc[((test_df.Pclass==1)|(test_df.Pclass==2))&(test_df.Age<11),'Survived'] = 1
test_df
```

```python
def random_binary(y):
    return np.random.randint(2)

condition = (((test_df.Pclass == 1) | (test_df.Pclass == 2)) &
             ((test_df.Age > 11) & (test_df.Age < 60)) &
             (test_df.Sex == 'male'))

test_df.loc[condition, 'Survived'] = test_df.loc[condition, 'Survived'].apply(random_binary)
test_df
```

```python
pd.options.mode.copy_on_write = True
def chkmr(a):
    if 'Mr.' in a:
        return a
        
#train_df[train_df.Name==train_df.Name.apply(chkmr).dropna()]
test_df.iloc[(test_df.Name.apply(chkmr).dropna().index)].Survived=0
test_df
```
