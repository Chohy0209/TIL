```python
#사용자와 컴퓨터는 각각 1~10 열장의 카드를 가지고 있으며 매 판 랜덤한 한장의 카드를 뽑아 진행하고 이미 뽑은 카드는 게임 한바퀴가 돌때까지 다시 나오지 않음
#컴퓨터의 숫자만을 보고 승패를 예측하고 각각의 상황에 따른 승점을 얻는 방식 매 라운드가 끝나기 전까지는 자신의 카드는 확인 불가.
#사용자가 승리를 예측하고 실제 승리시 2점 획득, 무승부시 컴퓨터와 사용자 모두 1점씩 획득 패배시 컴퓨터가 2점 획득
#사용자가 패배를 예측거나 승 패 외에 다른 말을 입력할시 그 즉시 그 라운드가 종료되며 컴퓨터가 1점 획득
#각 라운드가 끝나면 무슨카드가 뽑혔는지 알 수 있음 단 패를 예측시 자신의 카드 오픈 x
#1~10장의 카드를 모두 사용해 10라운드를 진행하면 10판의 합산 점수로 최종 승패 결정

import random

user_use = []  # 유저가 사용한 카드
com_use = []  # 컴퓨터가 사용한 카드
user_score = 0
com_score = 0


def indian():
    global user_score, com_score, user_use, com_use

    while True:
        user_set1 = random.randint(1, 10)  # 유저의 카드 한개 선정
        if user_set1 not in user_use:  # 1~10 까지의 카드 한 세트에서 게임을 진행 하기 위해 난수가 겹치지 않은경우 게임 실행
            com_set1 = random.randint(1, 10)
            if com_set1 not in com_use:  # 컴퓨터의 카드 선정
                print(com_set1)
                user_cho = input("승 패를 예측하세요")  # 유저가 승 패를 예측하고 승리에 예측시 높은 숫자를 보유한 쪽이 승리하며 2점 획득 패배 예측시 컴퓨터에게만 기본 1점 부여
                com_use.append(com_set1)
                user_use.append(user_set1) #이번판에 사용하는 유저와 컴퓨터 카드를 각각 사용한 카드 리스트에 추가
                if user_cho == '승':  # 유저가 승리 예측시
                    if user_set1 > com_set1:
                        print('승리하셨습니다 2점을 얻습니다 당신의 카드는', user_set1, '이었습니다')
                        user_score += 2
                        if len(user_use) == 10:  # 카드 10장 한세트를 기본으로 하기위해 사용한 카드를 넣은 리스트의 길이가 10일때 멈춤
                            if user_score > com_score:
                                print("당신의 점수는", user_score, '입니다')
                                print("computer의 점수는", com_score, '입니다')
                                print('승리했습니다')
                                break
                            elif user_score == com_score:
                                print("당신의 점수는", user_score, '입니다')
                                print("computer의 점수는", com_score, '입니다')
                                print('비겼습니다')
                                break
                            else:
                                print("당신의 점수는", user_score, '입니다')
                                print("computer의 점수는", com_score, '입니다')
                                print('졌습니다')
                                break

                    elif user_set1 == com_set1:
                        print('무승부 입니다.1점을 획득합니다.당신의 카드는', user_set1, '이었습니다')
                        user_score += 1
                        com_score += 1
                        if len(user_use) == 10:
                            if user_score > com_score:
                                print("당신의 점수는", user_score, '입니다')
                                print("computer의 점수는", com_score, '입니다')
                                print('승리했습니다')
                                break
                            elif user_score == com_score:
                                print("당신의 점수는", user_score, '입니다')
                                print("computer의 점수는", com_score, '입니다')
                                print('비겼습니다')
                                break
                            else:
                                print("당신의 점수는", user_score, '입니다')
                                print("computer의 점수는", com_score, '입니다')
                                print('졌습니다')
                                break

                    else:
                        print('졌습니다. 점수를 획득하지 못하였습니다.당신의 카드는', user_set1, '이었습니다')
                        com_score += 2
                        if len(user_use) == 10:
                            if user_score > com_score:
                                print("당신의 점수는", user_score, '입니다')
                                print("computer의 점수는", com_score, '입니다')
                                print('승리했습니다')
                            elif user_score == com_score:
                                print("당신의 점수는", user_score, '입니다')
                                print("computer의 점수는", com_score, '입니다')
                                print('비겼습니다')
                            else:
                                print("당신의 점수는", user_score, '입니다')
                                print("computer의 점수는", com_score, '입니다')
                                print('졌습니다')

                            break
                        else:
                            continue
                elif user_cho == '패':
                    print("패를 선택하셨으므로 com에게만 1점이 주어집니다")
                    com_score += 1

                else:
                    print("승 패만을 기입하세요 다른 언어를 사용시 패 처리가 됩니다.")
                    com_score+ = 1
        else:

            continue


indian()
print(user_use,com_use)
```
